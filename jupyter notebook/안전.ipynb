```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정 (맥북용)
rc('font', family='AppleGothic')

plt.rcParams['axes.unicode_minus'] = False



# '생활방범'인 데이터 추출
lifesafety_df = df[df['설치목적구분'] == '생활방범']

# '어린이보호'인 데이터 추출
childprotection_df = df[df['설치목적구분'] == '어린이보호']

# 개수를 기준으로 내림차순 정렬
lifesafety_sorted_df = lifesafety_df.sort_values(by='개수', ascending=False).head(15)
childprotection_sorted_df = childprotection_df.sort_values(by='개수', ascending=False).head(15)

# 시각화 크기 조정
plt.figure(figsize=(10, 10))

# x축과 y축을 바꿔서 시각화
plt.barh(lifesafety_sorted_df['관리기관명'], lifesafety_sorted_df['개수'], color='skyblue', label='생활방범')
plt.barh(childprotection_sorted_df['관리기관명'], childprotection_sorted_df['개수'], color='lightgreen', label='어린이보호')

plt.xlabel('개수')
plt.ylabel('관리기관명')
plt.title('생활방범 및 어린이보호 데이터 시각화')

# x축 눈금 레이블 90도 회전
plt.xticks(rotation=90)

# 범례 표시
plt.legend()
plt.savefig('안전')

# 그래프 표시
plt.show()






```
