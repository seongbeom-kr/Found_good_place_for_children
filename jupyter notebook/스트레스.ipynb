```python
import pandas as pd
```

```python
df = pd.read_excel('data/DUST/서울.xlsx')
df.columns
```

```python
df = pd.DataFrame(columns=df.columns)
df
```

```python
city = ['서울','울산','인천','부산','대구','대전','광주']
for i in city:
    df1 = pd.read_excel('data/DUST/{}.xlsx'.format(i))
    df = pd.concat([df,df1], axis = 0)

df
```

```python
df['측정소명'] = df['측정소명'].str.extract(r'\[(.*?)]')[0]
df
```

```python
df[['도시', '구']] = df['측정소명'].str.split(' ', n=1, expand=True)
df
```

```python
df.drop('측정소명', axis = 1, inplace = True)
```

```python
city = df[['도시','구']]
df = df.drop(['도시','구'],axis = 1)
df = df.apply(pd.to_numeric, errors='coerce')

df = pd.concat([city,df], axis = 1)


df = df.fillna(df.mean(axis = 1))
df = df.groupby(['도시','구']).mean()
```

```python
df['mean'] = df.iloc[:, 2:].sum(axis=1) / 31
```

```python
df.info()
```

```python
df_sorted = df.sort_values(by='mean', ascending=False).reset_index()
```

```python
df_sorted[['도시','구','mean']] # 웅진군, 달성군, 기장군, 울주군
```

```python
df_sorted = df_sorted[['도시','구','mean']]
```

```python
df_sorted =df_sorted[df_sorted['구']!='옹진군']
df_sorted =df_sorted[df_sorted['구']!='달성군']
df_sorted =df_sorted[df_sorted['구']!='기장군']
df_sorted =df_sorted[df_sorted['구']!='울주군']
```

```python
city = ['인천', '서울', '대전', '부산', '대구', '광주', '울산']
for i in city:
    if i == '서울':
        name = '특별시'
    else:
        name = '광역시'
    
    df_sorted['도시'][df_sorted['도시']==i] = i + name

```

```python
df_sorted['지역명'] = df_sorted['도시']+" "+df_sorted['구']
```

```python
df_sorted.drop(['도시', '구'], axis = 1, inplace = True)
```

```python
df_sorted.to_excel('스트레스 지수.xlsx')
```

```python
df_sorted = df_sorted.sort_values(by = 'mean')
```

```python
df_sorted.to_excel('스트레스 지수.xlsx')
```
